# Steps in modelling

```python
import tensorflow as tf
```

```python
# set random seed
tf.random.set_seed(42)

# 1. Create model
model = tf.keras.Sequential([
  tf.keras.Input(shape=(2,)),
  tf.keras.layers.Dense(5, activation='relu'),
  tf.keras.layers.Dense(1, activation='softmax')
])

# 2. Compile the model
model.compile(loss=tf.keras.losses.CategoricalCrossentropy( ),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])

# 3. Fit the model
model.fit(X, y, epochs=5)
```

```python
import numpy as np
```

```python
#Create function to visualize the model's predictions
def plot_decision_boundary(model, X, y):
  """
  Takes in features X and labels y,
  creates a meshgrid of the different X values,
  makes predictions across the meshgrid.
  Plot the predictions as well as a line between zones, where each unique class falls
  """
  # define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() -0.1, X[:, 0].max()+0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 0].max()+0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min,y_max, 100))
  
  # Create X values
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2d arrays together

  # Make predicitons
  y_pred = model.predict(x_in)

  # Check for multi-class
  if len(y_pred[0]) > 1:
    print("Doing multiclass classification")
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("Doing binary claffication")
    y_pred = np.round(y_pred).reshape(xx.shape)

  # Plot the decision boundary
  plt.contourf(xx,yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:,0], X[:,1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.xlim(yy.min(), yy.max())
```

```python
# Check the predictions the model is making
plot_decision_boundary(model=model_2, X=X, y=y)
```

```python
  # How does the model do with regresssion data
  tf.random.set_seed(42)

  # create regression data
  X_reg = tf.range(0, 1000, 5)
  y_reg = tf.range(100, 1100, 5) 

X_reg_train = X_reg[:150]
X_reg_test = X_reg[150:]
y_reg_train = y_reg[:150]
y_reg_test = y_reg[150:]

```

```python
# set random seed
tf.random.set_seed(42)

# 1. Create the model
model_reg = tf.keras.Sequential([
  tf.keras.layers.Dense(100),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1)
])

# 2. Compile the model
model_reg.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['mae'])

# 3. Fit the model
model_reg.fit(X_reg_train, y_reg_train, epochs=100)
```

```python
y_reg_preds = model_reg.predict(X_reg_test)

# plot the models predictions against regression data
plt.figure(figsize=(10,7))
plt.scatter(X_reg_train, y_reg_train, c='b', label='Training data')
plt.scatter(X_reg_test, y_reg_test, c='g', label='Test data')
plt.scatter(X_reg_test, y_reg_preds, c='r', label='Predictions')
plt.legend();
```

# Evaluating and improving the classification model

```python
# Creating training and testing sets
X_train, y_train = X[:800], y[:800]
X_test, y_test = X[800:], y[800:]
X_test.shape, X_train.shape, y_test.shape, y_train.shape
```

```python
# Create another model

# Set random seed
tf.random.set_seed(42)

# 1. Create the model
model_8 = tf.keras.Sequential([
  tf.keras.layers.Dense(4, activation='relu'),
  tf.keras.layers.Dense(4, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

# 2. Compile the model
model_8.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(lr=0.01),
                metrics=['accuracy'])

# 3. Fit the model
history = model_8.fit(X_train, y_train, epochs=25)
```

```python
model_8.evaluate(X_test, y_test)
```

```python
plot_decision_boundary(model=model_8, X=X_test, y=y_test)
```
